### Challenge 1: Largest Palindrome

```python
# Option 1: First attempt

# Function will take a string as an argument and determine if it is Palindrome.
# Return True if Palindrome else Return False
def isPalindrome(myNumStr):
    
    # Check if argument is a string
    if not isinstance(myNumStr, str):
        return False
    length = len(myNumStr)
    
    #For loop that runs to mid point of string array and check i and len-i-1 element to be same.
    # At any point if they are not the same, return False
    # If the loop finishes, then its a palindrome
    for i in range(0, length//2):
        if (myNumStr[i] != myNumStr[length - i -1]):
            return False
    return True


# Function that returns the largest palindrome made by mulitplying 2 3-digit numbers
def largest3NumberPalindrome():
    # Set to the largest 3 digit number
    maxdigit3 = 999
    largestPalindrome = 0
    
    # Start from largest 3 digit number and count down to lowest 3 digit number 100. 
    # Since second argument of range is exclusive, we can go till 999//10 = 99.
    for i in range(maxdigit3, maxdigit3 // 10, -1):
        #Second for loop to keep counting down from 999 to 100 (inclusive)
        for j in range(maxdigit3, maxdigit3 // 10, -1):
            # Mulitple the numbers 
            temp = i * j
            # Check if temp=i *j is a palindrome and if temp is greater than the previously set largestPalindrome
            # Pass to isPalindrome string value of the temp as temp is integer
            if isPalindrome(str(temp)) and temp > largestPalindrome:
                # reset largestPalindrome to temp
                largestPalindrome = temp
    return largestPalindrome
    

print(largest3NumberPalindrome())
```


### Challenge 2: Summation of Primes
The sum of the primes below 10 is 2 + 3 + 5 + 7 = 17. Find the sum of all the primes below 2,000. Afterward, write a brief explanation walking through your code's logic in markdown.

```python
# Returns True if number is Prime, False if no
def isPrime(myNum):
    if myNum <= 1:
        return False
    
    # Check if number is divisible by any number except for 1 and itself
    for i in range(2, 1 + myNum//2):
        if myNum % i == 0:
            return False
        
    return True

# Returns the sum of all Prime numbers between 1 and 2000
def sumOfAllPrimes():
    sumPrime = 0
    # Loop from 0 to < 2000 and if Prime, sum the numbers
    for i in range(0, 2000):
        if isPrime(i):
            sumPrime += i
    return sumPrime

sumOfAllPrimes()
```

### Challenge 3: Multiples of 3 and 5
If we list all of the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6, and 9. The sum of these multiples is 23. Find the sum of all the multiples of 3 and 5 below 1,000. Afterward, write a brief explanation walking through your code's logic in markdown.

```python
# Returns True if number is divisible by both 3 and 5
def isMulitpleOf3And5(myNum):
    if myNum % 5 == 0 and myNum % 3 == 0:
        return True
    return False

# Loop through 1 to 1000 and if divisible by 3 and 5, sum the numbers
def sumOfNumbersDivisbleBy3And5():
    sum3And5 = 0
    # for loop from 1 to 1000
    for i in range(1,1000):
        # check if current number is divisible by 3 and 5, then add the number to existing sum
        if isMulitpleOf3And5(i):
            sum3And5 += i
    return sum3And5

sumOfNumbersDivisbleBy3And5()
```

```python
# Returns the size of compressed string
def countComression(myStr):
    # If string is null or empty, return 0
    if not myStr:
        return 0
    else:
        # set to first character
        prevChar = myStr[0]
        size = 0
        cnt = 1
        
        # loop through from second character to end of string
        for i in range(1, len(myStr)):
            # Check if current character equals previous character
            if myStr[i] == prevChar:
                # increment count
                cnt += 1
            else:
                # set previous character to current
                prevChar = myStr[i]
                # set size to char + char occupied in the string to store the number of occurances of the string
                size += 1 + len(str(cnt))
                cnt = 1
        # Last character
        # set size to char + char occupied in the string to store the number of occurances of the string 
        size += 1 + len(str(cnt))
    return size


# Case insensitive string compressor. Returns the lower case count
def stringCompressor(myStr):
    # Convert all characters to lowercase
    myStr = myStr.lower()
    # if size of the compressed string is greater than or equal to original string, return the original string
    if countComression(myStr) >= len(myStr):
        return myStr
    
    # Define a string for storing the compressed string
    newStr = ""
    # Set to first character
    prevChar = myStr[0]
    cnt = 1
    
    # loop through from second character to end of string
    for i in range(1, len(myStr)):
        # Check if current character equals previous character and if yes, increase cnt
        if myStr[i] == prevChar:
            cnt += 1
        #else append the string and the count of the repeatable string to newStr
        else:
            newStr = newStr + prevChar + str(cnt)
            prevChar = myStr[i]
            cnt = 1
    
    return newStr + prevChar + str(cnt)

stringCompressor("aabcccccaaa")
```

### Challenge 5: CSV Reader
Write a function `read_csv` that opens a CSV file and returns a list of named tuples. 

+ Read the [namedtuple docs](https://docs.python.org/3/library/collections.html#collections.namedtuple).
+ Do not use the Python CSV reader standard library.
+ Assume the first row of the CSV file is always a header row. 
+ Assume the CSV file does not contain any quoted fields.
+ Assume no fields are missing or have invalid values.
+ All returned fields can be left as strings.

Optional challenges:

+ _Optional:_ Implement the `dtypes` parameter. This dictionary specifies the data type that each CSV field should be converted to. For example, `int` would indicate the field should be converted from a string to an integer. If a field is not named, leave it as a string. See the dictionary in the below starter code for an example.

_Hint:_ The final output will look like:

[Row(name='Lamusi A', sport='Judo', age='23', height='170', weight='60', country="People's Republic of China", sex='M', gold='0', silver='0', bronze='0', total='0'), 

Row(name='A G Kruger', sport='Athletics', age='33', height='193', weight='125', country='United States of America', sex='M', gold='0', silver='0', bronze='0', total='0'),

...]

```python
from collections import namedtuple

def read_csv(filename, dtypes={}, encoding='utf-8'):
    """ Returns a list of namedtuples. 
    The fields in each namedtuple are specified by the header row."""
    rows = []
    
    # Your code here!
    # Open the files and read the lines into myLines
    fi = open(filename, "r")
    myLines = fi.readlines()
    
    #split the list into header and data
    header = myLines[0]
    data = myLines[1::]
    # Instantiate a namedtuple Row with fielnames as header var
    Row = namedtuple('Row', header)
    
    # for each line in data, create a named tuple and add to rows array
    for l in data:
        # remove newline character and split using commas
        l1 = l.replace('\n','').split(',')
        # Use the _make method of named tuple to create a new named tuple
        r = Row._make(l1)
        
        
        
        # Append the row to rows array
        rows.append(r)
    
    return rows


# For the challenge problem only
dtypes = {'age': int, 'height': int, 'weight': int, 
          'gold': int, 'silver': int, 'bronze': int, 'total': int}

read_csv('./datasets/athletes-unquoted.csv', dtypes=dtypes)
```

### *BONUS* Challenge: FizzBuzz
Write a program that prints all of the numbers from 1 to 100. For multiples of 3, instead of the number, print "Fizz;" for multiples of 5, print "Buzz." For numbers that are multiples of both 3 and 5, print "FizzBuzz." Afterward, write a brief explanation walking through your code's logic in markdown.

```python
def printFizzBuzz():
    # loop through 1 to 100
    for i in range(1, 100):
        # If divisible by both 3 and 5, then print FizzBuzz
        if (i % 3 == 0) and (i % 5 == 0):
            print("FizzBuzz")
        # If divisible only by 3, then print Fizz
        elif (i % 3 == 0):
            print("Fizz")
        # If divisible only by 5, then print Buzz
        elif i % 5 == 0:
            print("Buzz")
        # Else print the number
        else:
            print(i)

printFizzBuzz()       
```
